```python
# Function to safely extract director and stars
def extract_director_and_stars(df):
    # Ensure the 'STARS' column is of string type
    df['STARS'] = df['STARS'].astype(str)

    # Create new columns for Director and Stars
    df['Director'] = ''
    df['Stars_new'] = ''

    for index, row in df.iterrows():
        stars_string = row['STARS']
        if '|' in stars_string:
            parts = stars_string.split('|')
            director = parts[0].replace('Director:', '').strip()
            stars = parts[1].replace('Stars:', '').strip()
            df.loc[index, 'Director'] = director
            df.loc[index, 'Stars_new'] = stars
        else:
            df.loc[index, 'Director'] = 'Unavailable'
            df.loc[index, 'Stars_new'] = stars_string.replace('Stars:', '').strip()

    # Drop the old 'STARS' column and rename 'Stars_new'
    df.drop('STARS', axis=1, inplace=True)
    df.rename(columns={'Stars_new': 'STARS'}, inplace=True)

    return df

df = extract_director_and_stars(df)


# Reorder the columns to place 'Director' next to 'STARS'
cols = list(df.columns)
cols.insert(6, cols.pop(cols.index('Director')))
df = df[cols]

# Display the first few rows to show the result
display(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select only the numeric columns for correlation
numeric_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation = numeric_df.corr()

# Create the heatmap
plt.figure(figsize=(5,4))
sns.heatmap(correlation, annot=True, cmap='Blues')
plt.show()
```
